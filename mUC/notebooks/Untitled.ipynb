```python
height_df = df.query('Test == "body height"')

mask_needs_height_imputation = (
     height_df['TestResultCleaned'].isna() & height_df['TestResult'].notna()
     )

imputed_heights = height_df.loc[mask_needs_height_imputation, 'TestResult'].apply(
     lambda x: x * 2.54 if 55 <= x <= 80  # Convert to cm if likely inches
     else x if x >= 120  # Keep as is if likely cm
     else None  # Leave as null if implausible/ambiguous
     )

height_df.loc[mask_needs_height_imputation, 'TestResultCleaned'] = imputed_heights
```

```python
height_df = (
    height_df
    .groupby('PatientID')['TestResultCleaned'].mean()
    .reset_index()
    .assign(TestResultCleaned = lambda x: x['TestResultCleaned']/100)
    .rename(columns = {'TestResultCleaned': 'height'})
)
```

```python
weight_df = pd.merge(weight_df, height_df, on = 'PatientID', how = 'left')
```
